```python
import pandas as pd
import sqlite3
import os

conn = sqlite3.connect("../data/db/bluestock_mf.db")

files = os.listdir("../data/processed")

for file in files:
    
    if file.endswith(".csv"):
        
        df = pd.read_csv(
            f"../data/processed/{file}"
        )
        
        table_name = (
            file
            .replace("_clean.csv", "")
        )
        
        df.to_sql(
            table_name,
            conn,
            if_exists="replace",
            index=False
        )
        
        print(
            f"{table_name} loaded successfully"
        )

print("\nAll tables loaded!")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
for table in [
    "aum_by_fund_house",
    "benchmark_indices",
    "category_inflows",
    "industry_folio_count",
    "investor_transactions",
    "monthly_sip_inflows",
    "nav_history",
    "portfolio_holdings",
    "scheme_performance"
]:
    
    query = f"""
    SELECT COUNT(*) as rows
    FROM {table}
    """
    
    print(table)
    print(pd.read_sql(query, conn))
    print("-"*50)
```

```python
query = """
SELECT
    scheme_name,
    aum_crore
FROM scheme_performance
ORDER BY aum_crore DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    AVG(nav) AS avg_nav
FROM nav_history;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    risk_grade,
    COUNT(*) AS fund_count
FROM scheme_performance
GROUP BY risk_grade;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    category,
    ROUND(AVG(expense_ratio_pct), 2) AS avg_expense_ratio
FROM scheme_performance
GROUP BY category
ORDER BY avg_expense_ratio DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    category,
    SUM(net_inflow_crore) AS total_inflow
FROM category_inflows
GROUP BY category
ORDER BY total_inflow DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    SUM(sip_inflow_crore) AS total_sip_inflow
FROM monthly_sip_inflows;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    AVG(sharpe_ratio) AS avg_sharpe
FROM scheme_performance;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    scheme_name,
    sharpe_ratio
FROM scheme_performance
ORDER BY sharpe_ratio DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    fund_house,
    ROUND(AVG(aum_crore), 2) AS avg_aum
FROM scheme_performance
GROUP BY fund_house
ORDER BY avg_aum DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    transaction_type,
    COUNT(*) AS total_transactions
FROM investor_transactions
GROUP BY transaction_type;
"""

pd.read_sql(query, conn)
```
